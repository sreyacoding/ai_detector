```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import re
import string
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, classification_report
import tensorflow as tf
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
import nltk
import pickle
import warnings
```

```python

# Suppress warnings and download NLTK data
warnings.filterwarnings("ignore")
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Load Dataset
data = pd.read_csv('Training_Essay_Data.csv')
```

```python
# Initial Data Inspection
print("\n--- Data Overview ---")
print("Shape of the dataset:", data.shape)
print("Column names:", data.columns)
print("\nFirst 5 Rows of the Data:\n", data.head())

print("\nDataset Information:")
data.info()  # Check column types, memory usage, and non-null counts
```

```python

print("\nStatistical Summary:")
print(data.describe())  # Check basic statistics for numerical columns

print("\nChecking for Missing Values:\n", data.isnull().sum())

# Checking for duplicates
print("\nNumber of Duplicate Entries:", data.duplicated().sum())
```

```python
# Data Exploration
print("\nClass Distribution:\n", data['generated'].value_counts())

plt.figure(figsize=(6, 4))
sns.countplot(x='generated', data=data, palette='viridis')
plt.title('Class Distribution: AI-Generated vs Human-Written')
plt.show()
```

```python
# Text Length Analysis
data['text_length'] = data['text'].apply(lambda x: len(x.split()))
print("\nText Length Summary:\n", data['text_length'].describe())

plt.figure(figsize=(10, 6))
sns.histplot(data['text_length'], bins=30, kde=True, color='purple')
plt.title('Text Length Distribution (Word Count)')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()
```

```python
# Stopword Usage
def count_stopwords(text):
    return len([word for word in text.split() if word in stopwords.words('english')])

data['stopword_count'] = data['text'].apply(count_stopwords)

print("\nSummary of Stopword Usage:")
print(data['stopword_count'].describe())

plt.figure(figsize=(10, 6))
sns.boxplot(x='generated', y='stopword_count', data=data)
plt.title('Distribution of Stopword Usage by Class')
plt.show()
```

```python
# WordClouds for AI and Human Texts
ai_generated_text = " ".join(data[data['generated'] == 1]['text'].tolist())
human_written_text = " ".join(data[data['generated'] == 0]['text'].tolist())

wordcloud_ai = WordCloud(width=800, height=400, background_color='white').generate(ai_generated_text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_ai, interpolation='bilinear')
plt.title('WordCloud for AI-Generated Essays')
plt.axis('off')
plt.show()

wordcloud_human = WordCloud(width=800, height=400, background_color='white').generate(human_written_text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_human, interpolation='bilinear')
plt.title('WordCloud for Human-Written Essays')
plt.axis('off')
plt.show()
```

```python
# Checking for Imbalanced Classes
print("\nClass Imbalance Check:")
print(data['generated'].value_counts(normalize=True))

# Data Cleaning and Preprocessing
# Replace NaNs with empty strings
data['text'].fillna('', inplace=True)  
```

```python
# Text Cleaning Function
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)  # Remove punctuation
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    text = " ".join([lemmatizer.lemmatize(word) for word in text.split() if word not in stopwords.words('english')])
    return text

data['text_cleaned'] = data['text'].apply(clean_text)
```

```python
# Text Length After Cleaning
data['cleaned_length'] = data['text_cleaned'].apply(lambda x: len(x.split()))
print("\nCleaned Text Length Summary:\n", data[['text_length', 'cleaned_length']].describe())
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(data['text_cleaned'], data['generated'], test_size=0.2, random_state=42)

print("\nTrain-Test Split:")
print(f"Training Data: {X_train.shape[0]} samples")
print(f"Test Data: {X_test.shape[0]} samples")
```

```python
# Vectorization using TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
# Step 8: Train Machine Learning Models

# Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train_tfidf, y_train)

# Get probabilities and predictions
lr_prob = lr_model.predict_proba(X_test_tfidf)  # Probabilities for each class
lr_pred = lr_model.predict(X_test_tfidf)       # Predicted class labels

# Random Forest
rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(X_train_tfidf, y_train)

# Get probabilities and predictions
rf_prob = rf_model.predict_proba(X_test_tfidf)  # Probabilities for each class
rf_pred = rf_model.predict(X_test_tfidf)        # Predicted class labels

# SVM
svm_model = SVC(probability=True)  # Enable probability predictions for svm
svm_model.fit(X_train_tfidf, y_train)

# Get probabilities and predictions
svm_prob = svm_model.predict_proba(X_test_tfidf)  # Probabilities for each class
svm_pred = svm_model.predict(X_test_tfidf)        # Predicted class labels

```

```python
# Fine-tuning BERT Model

# Initialize BERT Tokenizer and Model
bert_tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')
bert_model = TFAutoModelForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=2)

# Tokenize Data for BERT
def tokenize_bert(sentences):
    return bert_tokenizer(sentences, padding=True, truncation=True, max_length=512, return_tensors='tf')

X_train_bert = tokenize_bert(X_train.tolist())
X_test_bert = tokenize_bert(X_test.tolist())

# Train BERT Model
bert_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=2e-5),
                   loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                   metrics=['accuracy'])

bert_model.fit(X_train_bert['input_ids'], y_train, epochs=2, batch_size=8, validation_split=0.2)
```

```python
bert_pred = bert_model.predict(X_test_bert['input_ids'])
bert_pred = np.argmax(bert_pred.logits, axis=1)
```

```python
# Define the evaluation function
def evaluate_model(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    return accuracy, precision, recall, f1
```

```python
# Evaluate each model and store their metrics
lr_metrics = evaluate_model(y_test, lr_pred)
rf_metrics = evaluate_model(y_test, rf_pred)
svm_metrics = evaluate_model(y_test, svm_pred)
bert_metrics = evaluate_model(y_test, bert_pred)

# Print evaluation metrics for each model
print("Logistic Regression Metrics:", lr_metrics)
print("Random Forest Metrics:", rf_metrics)
print("SVM Metrics:", svm_metrics)
print("BERT Metrics:", bert_metrics)
```

```python
# Store models and their F1-scores
model_scores = {
    "Logistic Regression": (lr_model, lr_metrics[3]),  
    "Random Forest": (rf_model, rf_metrics[3]),
    "SVM": (svm_model, svm_metrics[3]),
    "BERT": (bert_model, bert_metrics[3]),
}

# Select the model with the highest F1-score
best_model_name, (final_model, best_f1_score) = max(model_scores.items(), key=lambda x: x[1][1])
print(f"\nBest Model: {best_model_name} with F1-Score: {best_f1_score:.2f}")
```

```python
# Save the best model to a pickle file
with open('final_model.pkl', 'wb') as f:
    pickle.dump(final_model, f)
```

```python
# Load the saved model
with open('final_model.pkl', 'rb') as f:
    loaded_model = pickle.load(f)
```

```python
# Example usage: Predict whether the text is AI-Generated or Human-Written
example_text = "Global warming refers to the long-term increase in Earth's average surface temperature due to human activities, particularly the burning of fossil fuels like coal, oil, and natural gas. These activities release greenhouse gases such as carbon dioxide and methane, which trap heat in the atmosphere. The effects of global warming include rising sea levels, melting glaciers, extreme weather events, and disruptions to ecosystems and biodiversity. If not addressed, it could lead to food and water shortages, displacement of populations, and increased health risks. Mitigating global warming requires global efforts in reducing emissions, transitioning to renewable energy, and adopting sustainable practices."

# Assuming clean_text() is defined to preprocess the text
example_cleaned = clean_text(example_text)  

# Assuming tfidf_vectorizer is already fitted and available
example_vectorized = tfidf_vectorizer.transform([example_cleaned])  

# Predict using the loaded final model
prediction = loaded_model.predict(example_vectorized)

# Output the prediction result
print("Prediction for the example text:", "AI-Generated" if prediction[0] == 1 else "Human-Written")
```
